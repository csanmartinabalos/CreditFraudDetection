## Ejecución de modelo de prueba con los datos tal y como estan

```python
X = data.drop(['Class'], axis=1)
Y = data['Class']
print(X.shape)
print(Y.shape)

#guardamos los datos en forma de array de numpy para su utilización
xData = X.values
yData = Y.values
```

DIVISION DE ENTRENAMIENTO Y PRUEBA

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, Y_train, Y_valid = train_test_split(xData, yData, train_size=0.8, test_size=0.2, random_state=42)
```

GENERACIÓN DEL MODELO

```python
from sklearn.ensemble import RandomForestClassifier

model_RFC = RandomForestClassifier()
model_RFC.fit(X_train, Y_train)
Y_pred = model_RFC.predict(X_valid)
```

CONSTRUYENDO TODO TIPO DE PARAMETROS DE EVALUACION

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score, matthews_corrcoef
from sklearn.metrics import confusion_matrix, mean_absolute_percentage_error

n_outliers = len(fraud)
n_errors = (Y_pred != Y_valid).sum()

print('Modelo Random Forest Classifier')
acc = accuracy_score(Y_valid, Y_pred)
print('Accuracy:', acc)

pre_score = precision_score(Y_valid, Y_pred)
print('Precision score:', pre_score)

rec_score = recall_score(Y_valid, Y_pred)
print('Recall score:', rec_score)

f1 = f1_score(Y_valid, Y_pred)
print('F1 score:', f1)

mcc = matthews_corrcoef(Y_valid, Y_pred)
print('Matthews Correlation Coeficient score:', mcc)

mape = mean_absolute_percentage_error(Y_valid, Y_pred)
print('Mean absolute percentage error:', mape)
```

```python
conf_matrix = confusion_matrix(Y_valid, Y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='.2f')
```
